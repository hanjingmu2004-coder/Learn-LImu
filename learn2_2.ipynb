练习1 删除缺失值最多的列

```python
def clean_missing_most(df):
    missing_counts = df.isnull().sum()
    missing_counts_max=missing_counts.max()
    missing_counts_max_col=missing_counts[missing_counts==missing_counts_max].index.tolist()
    if missing_counts_max_col:
        col_to_remove = missing_counts_max_col[0]
        return df.drop(columns=[col_to_remove])
    else:
        return df
```

```python
data=clean_missing_most(data)
data
```
