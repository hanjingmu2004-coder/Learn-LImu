```python
import torch
from torch import nn 
from d2l import torch as d2l
def comp_conv2d(conv2d,X):
    X=X.reshape((1,1)+X.shape)#batch_size and channel_num
    Y=conv2d(X)
    return Y.reshape(Y.shape[2:])
conv2d=nn.Conv2d(1,1,kernel_size=3,padding=1)
X=torch.rand(size=(8,8))
comp_conv2d(conv2d,X)
```

```python
conv2d=nn.Conv2d(1,1,kernel_size=3,padding=1,stride=2)
comp_conv2d(conv2d,X)
```

```python
import torch
from d2l import torch as d2l
def corr2d_multi_in(X,K):
    return sum(d2l.corr2d(x,k) for x,k in zip(X,K))
```

```python
X=torch.rand(size=(2,3,3))
```

```python
K=torch.rand(size=(2,2,2))
K.shape
K
```

```python
corr2d_multi_in(X,K)
```

```python
def corr2d_multi_in_out(X,K):
    return torch.stack([corr2d_multi_in(X,k) for k in K],0)#based on 0_d stack the output
K=torch.stack((K,K+1,K+2),0)
K.shape
```

```python
corr2d_multi_in_out(X,K)
```

```python
def corr2d_multi_in_out_11(X,K):
    c_i,h,w=X.shape
    c_o=K.shape[0]
    X=X.reshape((c_i,h*w))
    K=K.reshape((c_o,c_i))
    Y=K@X
    return Y.reshape((c_o,h,w))
```

```python
X=torch.normal(0,1,(3,3,3))
K=torch.normal(0,1,(2,3,1,1))
Y1=corr2d_multi_in_out_11(X,K)
```
