```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. 定义交叉熵损失函数
def cross_entropy(y, y_hat):
    eps = 1e-7
    return -(y * np.log(y_hat + eps) + (1 - y) * np.log(1 - y_hat + eps))


y = np.linspace(0, 1, 100)
y_hat = np.linspace(0, 1, 100)
Y, Y_HAT = np.meshgrid(y, y_hat)
Z = cross_entropy(Y, Y_HAT)

# 3. 绘图
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# 绘制曲面
# cmap 使用 'viridis' 或 'magma' 可以清晰显示高度变化
surf = ax.plot_surface(Y, Y_HAT, Z, cmap='viridis', edgecolor='none', alpha=0.8)

# 设置坐标轴标签
ax.set_xlabel('True Label (y)')
ax.set_ylabel('Predicted Probability ($\hat{y}$)')
ax.set_zlabel('Cross Entropy Loss')
ax.set_title('3D Surface of Cross Entropy Loss')



# 添加颜色条
fig.colorbar(surf, shrink=0.5, aspect=10)

# 调整视角
ax.view_init(elev=30, azim=45)

plt.show()
```
